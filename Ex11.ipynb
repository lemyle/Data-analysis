## Ex11_2

```python
import numpy as np
import matplotlib.pyplot as plt
X = np.linspace(-2*np.pi, 2*np.pi, 20)
plt.plot(X, np.sin(X))
plt.plot(X, np.cos(X))
plt.plot(X, np.tan(X))
plt.legend(['sin', 'cos', 'tan'])
plt.show()
```

```python
# Vẽ đồ thị hàm x = y**2 và x = y trên cùng 1 trục toạ độ.
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-6, 7)
plt.plot(x, x**2)
plt.plot(x, x)
plt.legend(['y=x**2', 'y=x'])
plt.show()
```

```python
# Tạo 1000 dữ liệu kiểu int ngẫu nhiêu gắn với 1000 ngày từ 09/02/2010. Dùng kiểu pd.Series. Tính cumsum() và vẽ đồ thị.
```

```python
import pandas as pd
D = np.random.randint(-9,10,1000)
df = pd.Series(data=D,  index=pd.date_range('2010/2/9', periods=1000))
```

```python
df = np.cumsum(df)
df.plot()
```

```python
%matplotlib inline
df = df.sort_values('2015')
df = df.dropna(0, how='all')
```
